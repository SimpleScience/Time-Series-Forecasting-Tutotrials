# Loading the Required Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Creating time series data

```python
# Creating a time range
time = pd.date_range(start='1/1/2024', periods=365, freq='D')

# Creating a simple time series data (base)
data = pd.Series(np.arange(len(time)), index=time, name='Base')

# Adding Trend: Linearly increasing values over time
trend = data * 0.1

# Adding Seasonality: Sinusoidal pattern to represent seasonal effect
seasonality = 10 + np.sin(np.linspace(0, 3 * np.pi, len(time))) * 10

# Adding Noise: Random noise
noise = np.random.normal(0, 2, len(time))

# Combine all components
combined = data + trend + seasonality + noise
```

# Plots of the Data

```python

# Plotting the components and the combined time series
plt.figure()
plt.plot(data, label='Base')
plt.title('Base Time Series')
plt.legend()


plt.figure()
plt.plot(trend, label='Trend', color='green')
plt.title('Trend')
plt.legend()

plt.figure()
plt.plot(seasonality, label='Seasonality', color='orange')
plt.title('Seasonality')
plt.legend()


plt.figure()
plt.plot(combined, label='Combined (Base + Trend + Seasonality + Noise)', color='red')
plt.title('Combined Time Series')
plt.legend()
```
